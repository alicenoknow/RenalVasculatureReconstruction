```python
from PIL import Image
from sklearn import preprocessing
from skimage import io, morphology
from skimage.filters import gaussian
from scipy.ndimage import distance_transform_edt

import numpy as np
import networkx as nx
```

```python
from matplotlib import pyplot as plt
```

```python
from skimage.color import rgb2gray
from skimage import util 
```

```python
I = Image.open('test2d.png').convert('RGB')
print(I.format)
print(I.size)
print(I.mode)
```

```python
img = np.asarray(I)
img_gray = rgb2gray(img)
# img_gauss = gaussian(img_gray, sigma=0.6)
```

```python
img_bin = preprocessing.Binarizer(threshold=0.34).transform(img_gray)
img_bin = util.invert(img_bin).astype(np.uint8)
```

```python
# img_gauss = morphology.area_opening(img_bin, area_threshold=60)
# img_gauss = morphology.area_closing(img_bin, area_threshold=10)
# img_gauss = morphology.binary_dilation(img_gauss)
img_gauss = img_bin
```

```python
plt.imshow(img_bin, cmap='gray')
plt.show()
```

```python
plt.imshow(img_gauss, cmap='gray')
plt.show()
```

```python
img_edt = distance_transform_edt(img_gauss)

bounds = int(np.ceil(np.max(img_edt)))
# img_gauss = np.pad(img_gauss, ((bounds,bounds),(bounds,bounds)), 'constant')
```

```python
plt.imshow(img_edt)
plt.show()
```

```python
bounds = int(np.ceil(np.max(img_edt)))

img_bin = np.pad(img_bin, ((bounds,bounds), (bounds,bounds)), 'constant')
```

```python
graph = get_data_graph(img_bin)
```

```python
nx.draw_networkx(graph, with_labels=True)
```

```python
nodes = list(graph.nodes)
```

```python
x = [point[1]-bounds for point in nodes]
y = [point[0]-bounds for point in nodes]
```

```python
plt.imshow(img_gauss)
plt.scatter(x,y)
plt.show()
```
